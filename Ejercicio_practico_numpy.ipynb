###**Ejercicio práctico:**

```python
import numpy as np

url = 'https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv'

columna = np.arange(1,6,1)
datos = np.loadtxt(url, delimiter=',',skiprows=1,usecols=columna)
datos_transpuestos = datos.T

print(datos_transpuestos)

print(f'Filas: {datos_transpuestos.shape[0]}, Columnas: {datos_transpuestos.shape[1]}')
print('\n')
print(f'Información sobre las Naranjas: \n\n Diametro: {datos_transpuestos[0]}\n Peso: {datos_transpuestos[1]}\n Rojo: {datos_transpuestos[2]}\n Verde: {datos_transpuestos[3]}\n Azul: {datos_transpuestos[4]}')
```

```python
naranjas = datos_transpuestos[0:2,:5000]
torojas = datos_transpuestos[0:2,5000:]

print('Naranjas:')
print(naranjas)
print('\n')
print('Toronjas')
print(torojas)
```

```python
import matplotlib.pyplot as plt

diametro_naranja = naranjas[0,:]
peso_naranja = naranjas[1,:]

print('Diametro Naranjas:')
print(diametro_naranja)
print('\n')
print('Peso Naranjas:')
print(peso_naranja)

plt.plot(diametro_naranja,peso_naranja)
```

```python
diametro_toronja = torojas[0,:]
peso_toronja = torojas[1,:]

print('Diametro Torojas:')
print(diametro_toronja)
print('\n')
print('Peso Toronja:')
print(peso_toronja)

plt.plot(diametro_toronja,peso_toronja)
```

Calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.

##*Formula de la recta:*
##**y = (a * x) + b**

**Naranjas**

```python
xn = diametro_naranja
yn = peso_naranja
n_n = np.size(xn)

a_n = (n_n*np.sum(xn*yn) - np.sum(xn) * np.sum(yn)) / (n_n * np.sum(xn**2) - np.sum(xn)**2)
b_n = np.mean(yn) - a_n * np.mean(xn)
```

```python
plt.plot(xn, yn)
plt.plot(xn, a_n*xn + b_n)
```

**Toronjas**

```python
xt = diametro_toronja
yt = peso_toronja
n_t = np.size(xt)

a_t = (n_t*np.sum(xt*yt) - np.sum(xt) * np.sum(yt)) / (n_t * np.sum(xt**2) - np.sum(xt)**2)
b_t = np.mean(yt) - a_t * np.mean(xt)
```

```python
plt.plot(xt, yt)
plt.plot(xt, a_t*xt + b_t)
```

**Naranjas**

```python
norma_naranjas = np.array([])

np.random.seed(84)
pendientes_naranjas = np.random.uniform(low=0.0, high=30.0, size=100)

for i in range(100):
  norma_naranjas = np.append(norma_naranjas, np.linalg.norm(yn - (pendientes_naranjas[i] * xn + 17)))

mejor_indice_naranjas = np.argmin(norma_naranjas)
mejor_pendiente_naranjas = pendientes_naranjas[mejor_indice_naranjas]
print(f'La mejor pendiente es: {mejor_pendiente_naranjas}')
```

```python
plt.plot(xn, yn)
plt.plot(xn, mejor_pendiente_naranjas*xn + b_n)
```

**Toronjas**

```python
norma_toronjas = np.array([])

np.random.seed(84)
pendientes_toronjas = np.random.uniform(low=0.0, high=30.0, size=100)

for i in range(100):
  norma_toronjas = np.append(norma_toronjas, np.linalg.norm(yt - (pendientes_toronjas[i] * xt + 17)))

mejor_indice_toronjas = np.argmin(norma_toronjas)
mejor_pendiente_toronjas = pendientes_toronjas[mejor_indice_toronjas]
print(f'La mejor pendiente es: {mejor_pendiente_toronjas}')
```

```python
plt.plot(xt, yt)
plt.plot(xt, mejor_pendiente_toronjas*xt + b_t)
```
